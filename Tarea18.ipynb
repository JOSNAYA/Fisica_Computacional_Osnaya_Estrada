# Diagrama de bifurcaci√≥n

```julia
function diagrama_bifurcacion(x,r,n)
     for i in 1:n                               #N-veces para obtener la curva
        x = r*x*(1-x)                          #Definimos a la curva logistica.
    end
    return x
end
;
```

```julia
#Intervalos de los valores de x y r
x=linspace(0,1,1000)
r=linspace(1,4,1000)
lista_xs=[]
lista_rs=[]
for i in 1:length(r)
    push!(lista_xs,diagrama_bifurcacion.(x,r[i],1000))
    push!(lista_rs,ones(length(x))*r[i])
end
```

```julia
xlabel("r"), ylabel("x"), title("Diagrama de Bifurcacion")
plot(lista_rs,lista_xs,"o",markersize=0.005)
;
```

# Exponente de Lyapunov

```julia
function Exponente_Lyapunov(x,n,r)
    l=0
   for i in 1:n
        x=r*x*(1-x)
        l=l+log(abs(r*(1-2*x)))
    end
    return l/n
end
;
```

```julia
title("Exponente de Lyapunov")
plot(r,Exponente_Lyapunov.(0.5,1000,r),"o",markersize=0.5)
title("Exponente de Lyapunov"), legend();
```
