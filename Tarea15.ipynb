# Probando método de Euler explicito

```julia
#Función de la tarea 11, ejercicio 4
h(x,t)=[x[2],-x[1]]
conjunto=herramientas.Euler(h,[2.0,3.0],0.0,10.0,0.1)
plot(conjunto[1],map(x->x[1],conjunto[2]),label="h=0.5")

t=linspace(0,10,100)
y=2cos.(t)+3sin.(t)
plot(t,y,label="Solucion")

legend(),grid("on"),show();
```

# Probando método de Euler implicito

```julia
#la función es del notebook Tarea12, con x0=1 t(0,8), con h=0.5
p(x,t)=-2.5*x;
```

```julia
conjunto1=herramientas.Euler_implicito(p,1.0,0.0,8.0,0.5)
```

```julia
plot(conjunto1[1],map(x->x[1],conjunto1[2]),label="Implicito")
#Solución exacta
t=linspace(0.0,8.0,300)
plot(t,exp.(-2.5*t),label=("Solución"))

title("Condicion inicial 1.0, h=0.5")
legend(),grid("on");
```

# Probando método de Runge Kutta de 4 orden

```julia
#la función es del notebook Tarea13 con condiciones iniciales [1,0]
r(x,t)=[x[2],-x[1]-x[2]]
```

```julia
conjunto2=herramientas.Runge_Kutta_4(r,[1.0,0.0],0.0,100.0,0.1)
```

```julia
plot(map(x->x[1],conjunto2[2]),map(x->x[2],conjunto2[2]))
title("Diagrama Fase. Usando Runge Kutta 4"),grid("on");
```
